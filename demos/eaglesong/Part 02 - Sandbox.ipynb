Aside from running bots in Telegram, you can also run them in Jupyter Notebooks! 


```python
from kaia.eaglesong.core import *
from kaia.eaglesong.drivers.ipython import *

def main():
    yield f'Hello, say anything and I will repeat. Or /start to reset.'
    while True:
        input = yield Listen()
        yield input

def run(function):
    interpreter = IPythonInterpreter(Automaton(function, None), IPythonChatModel())
    chat = IPythonChatWidget(interpreter)
    return chat.run()
    
run(main)
```

For your convenience, we will also load all the bots we saw in the previous part, so you could play with them

```python
from kaia.eaglesong.drivers.ipython import IPythonInterpreter, IPythonChatWidget, IPythonChatModel
import importlib
from yo_fluq_ds import *
import sys

def get_bot_from_path(path):
    module_name = 'demos.eaglesong.'+path.name.split('.')[0]
    spec = importlib.util.spec_from_file_location(module_name, path)
    module = importlib.util.module_from_spec(spec)
    sys.modules[module_name] = module
    spec.loader.exec_module(module)
    return module.bot

bots = (Query
        .folder('.')
        .where(lambda z: z.name.startswith('example_'))
        .to_dictionary(lambda z: z.name, lambda z: get_bot_from_path(z))
       )

def run_bot(bot):
    interpreter = IPythonInterpreter(bot.create_generic_automaton(None), IPythonChatModel())
    chat = IPythonChatWidget(interpreter, timer=bot.timer, bot_name = bot.name)
    return chat.run()

print(list(sorted(bots)))
run_bot(bots['example_07_menu_1.py'])
```
